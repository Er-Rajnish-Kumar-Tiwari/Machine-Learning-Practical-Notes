```python
df=pd.DataFrame(elastic_net_data["data"])
df=pd.get_dummies(df)
```

```python
X=df.drop("Salary",axis=1)
y=df["Salary"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
elastic=ElasticNet()
```

```python
parameters={'alpha': [0.001, 0.01, 0.1, 1, 10]}
```

```python
elasticv=GridSearchCV(elastic,parameters,scoring="neg_mean_squared_error",cv=5)
elasticv.fit(X_train,y_train)
```

```python
elastic_pre=elasticv.predict(X_test)
```

```python
r2=r2_score(elastic_pre,y_test)
```
