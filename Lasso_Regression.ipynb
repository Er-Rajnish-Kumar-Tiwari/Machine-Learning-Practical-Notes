```python
df=pd.DataFrame(lasso_salary_data["data"])
```

```python
df=pd.get_dummies(df)
```

```python
X=df.drop("Salary",axis=1)
y=df["Salary"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
lasso_reg=Lasso()
```

```python
parameters={'alpha': [0.001, 0.01, 0.1, 1, 10]}
```

```python
lassocv=GridSearchCV(lasso_reg,parameters,scoring="neg_mean_squared_error",cv=5)
lassocv.fit(X_train,y_train)
```

```python
print(lassocv.best_params_)
```

```python
print(lassocv.best_score_)
```

```python
lasso_pre=lassocv.predict(X_test)
```

```python
plt.scatter(y_test, lasso_pre)

plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.title("Actual vs Predicted")

plt.grid()
plt.show()
```
