```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing # this package give the prebuild dataset in our code 
```

```python
data = fetch_california_housing()
```

```python
data
```

```python
df=pd.DataFrame(data.data)
```

```python
df
```

```python
df.columns=data.feature_names
```

```python
df
```

```python
# Depedent and independent features X is independent and Y is depedenet
X=df
y=data.target
```

```python
# Train test splict 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Standrazing the data set 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
#bimport linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
reg=LinearRegression()
reg.fit(X_train, y_train)
```

```python
# Perform 10-fold cross-validation (negative MSE)
mse = cross_val_score(reg, X_train, y_train, cv=10, scoring='neg_mean_squared_error')
```

```python
np.mean(mse)
```

```python
# predict value
```

```python
redict_reg=reg.predict(X_test)
```

```python
redict_reg
```

```python
plt.plot(y_test, label="Actual")
plt.plot(redict_reg, label="Predicted")

plt.legend()
plt.title("Actual vs Predicted Comparison")

plt.show()
```
