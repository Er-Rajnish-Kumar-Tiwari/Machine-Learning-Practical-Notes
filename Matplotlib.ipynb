```python
data={
    "salary": [
  25000,
  32000,
  28000,
  45000,
  39000,
  52000,
  61000,
  47000,
  35000,
  29000,
  72000,
  68000,
  54000,
  31000,
  26000,
  48000,
  57000,
  63000,
  41000,
  36000
]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
# Line plot :- it use for single line show data in our graph
```

```python
plt.plot(df["salary"],color="red",marker="o",linestyle="--",linewidth="2")
plt.grid()
```

```python
# Historgram Representation 
```

```python
plt.hist(df["salary"],bins=5,color="blue")
plt.grid()
```

```python
# Boxplot Graph show data in box plot
```

```python
plt.boxplot(df["salary"])
plt.grid()
```

```python
df["dept"]=["HR","WEB","APP","Manager","CEO"]*4
```

```python
df
```

```python
# Piechart  show data in piechart 
```

```python
count=df["dept"].value_counts()
```

```python
count
```

```python
plt.pie(count,labels=count.index,autopct="%1.1f",explode=[0,0.2,0.1,0.3,0.4],shadow=True)
plt.show("equal")
```

```python
# Countplot  show graph on our 2 columns obejct or values
```

```python
plt.bar(count.index,count,color=["red","blue","green","black"])
plt.legend(df["dept"])
```

```python
df["age"]=[12,17,16,13,15]*4
```

```python
df
```

```python
# Scatter plot - show relationship between 2 numerical colums or values 
```

```python
plt.scatter(df["salary"],df["age"],color="red")
```

```python
df["exp"]=[1,1.5,2,2.5,3]*4
```

```python
df
```

```python
# Multivariate analysis - compare 3 numerical columns and data
```

```python
# Bubble plot 
```

```python
plt.scatter(df["salary"],df["age"],s=df["exp"]*10,color="red")
plt.legend(df)
```

```python
# print legend in our graph
```

```python
color = {
    "HR": "red",
    "WEB": "blue",
    "APP": "orange",
    "Manager": "black",
    "CEO": "green"
}
```

```python
for dept, col in color.items():
    df_dept = df[df["dept"] == dept]
    plt.scatter(df_dept["salary"], df_dept["age"], c=col, label=dept)

plt.legend()
plt.show()
```

```python
data2 = {
    "year": [2019, 2020, 2021, 2022, 2023],
    "sales": [50000, 60000, 75000, 80000, 90000],
    "profit": [8000, 9000, 12000, 15000, 18000],
    "expense": [42000, 51000, 63000, 65000, 72000]
}
```

```python
newdf=pd.DataFrame(data2)
```

```python
newdf
```

```python
plt.plot(newdf["year"],newdf["sales"],label="Sales")
plt.plot(newdf["year"],newdf["profit"],label="Profit")
plt.plot(newdf["year"],newdf["expense"],label="Expense")

plt.title("Finance Analysis")
plt.xlabel("year")
plt.ylabel("amount")
plt.legend()
plt.grid()
```
